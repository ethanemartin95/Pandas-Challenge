```python
# Dependencies and Setup
import pandas as pd
import os as os

# Change Path
csvdir = os.chdir('/Users/ethan_martin95/Documents/Georgia-Tech-Bootcamp-PULL/GT-ATL-DATA-PT-12-2019-U-C/Homework/04-Pandas/Instructions/HeroesOfPymoli/Resources')
# Grab File
csvpath = os.path.join('purchase_data.csv')                  

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csvpath)
purchase_data.head()
    
    
```

```python
# Display the total number of players

# Count unique screen names in SN Column
count_players = purchase_data["SN"].nunique()

# Create Dataframe of Data
total_players = pd.DataFrame({"Total Players": [count_players]}, index = ["Data"])
total_players
```

```python
# Run basic calculations to obtain number of unique items, average price, etc.
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

# Count unique Item IDs in Item ID Column
count_unique_items = purchase_data["Item ID"].nunique()

# Get Average price in Price Column
average_price = purchase_data["Price"].mean()

# Return count of elements in Purchase ID Column
count_purchase = purchase_data["Purchase ID"].count()

# Sum the entire Price Column
total_revenue = purchase_data["Price"].sum()

# Create Dataframe of data
summary_df = pd.DataFrame({"Number of Unique Items": [count_unique_items],
                           "Average Price": [average_price],
                           "Number of Purchases": [count_purchase],
                           "Total Revenue": [total_revenue]},
                            index = ["Data"])

# Reformat data to show currency
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)

summary_df
```

```python
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed

#create small dataframe for each gender that returns the unique count of screen names
male_df = purchase_data.loc[purchase_data["Gender"] == "Male",["Gender","SN"]].groupby("Gender").nunique("SN")
male_count = male_df.iloc[0,1]

female_df = purchase_data.loc[purchase_data["Gender"] == "Female",["Gender","SN"]].groupby("Gender").nunique("SN")
female_count = female_df.iloc[0,1]

o_gender_df = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed",["Gender","SN"]].groupby("Gender").nunique("SN")
o_gender_count = o_gender_df.iloc[0,1]

# Create dataframe of data
gender_summary_df = pd.DataFrame({"Total Counts": [male_count, female_count, o_gender_count],
                           "Percentage of Players": [male_count/count_players*100, female_count/count_players*100, o_gender_count/count_players*100]},
                            index = ["Male", "Female", "Other / Non-Disclosed"])

# Format columns for percentages
gender_summary_df["Percentage of Players"] = gender_summary_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

gender_summary_df
```

```python
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

# Sum Price by Gender in small dataframe
sum_price_gender_df = purchase_data[["Gender", "Price"]].groupby("Gender").sum()

# Count Purchases by Gender in small dataframe
count_purchase_gender_df = purchase_data[["Gender", "Purchase ID"]].groupby("Gender").count()

# Gather average price per purchase in small dataframe
average_price_gender_df = purchase_data[["Gender", "Price"]].groupby("Gender").mean()

# Get count of players by gender in small dataframe
count_players_gender_df = purchase_data[["Gender", "SN"]].groupby("Gender").nunique()

# Drop Gender Column
count_players2_gender_df = count_players_gender_df[["SN"]]

# Combine all small dataframes
gender_summary1_df = pd.merge(sum_price_gender_df, count_purchase_gender_df, how='inner', on=['Gender'])
gender_summary2_df = pd.merge(gender_summary1_df, average_price_gender_df, how="inner", on = ['Gender'])
gender_summary_df = pd.merge(gender_summary2_df, count_players2_gender_df, how= 'inner', on = ["Gender"])

# Create new column for average purchase price per person
gender_summary_df["Avg Total Purchase per Person"] = gender_summary_df["Price_x"]/gender_summary_df["SN"]

# Keep only needed columns
purchasing_gender_df = gender_summary_df[["Purchase ID", "Price_y", "Price_x", "Avg Total Purchase per Person"]]

# Rename Columns
purchasing_gender_df = purchasing_gender_df.rename(columns = {"Purchase ID":"Purchase Count",
                                                              "Price_y": "Average Purchase Price",
                                                              "Price_x": "Total Purchase Value"})

# Format columns for currency
purchasing_gender_df["Average Purchase Price"] = purchasing_gender_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchasing_gender_df["Total Purchase Value"] = purchasing_gender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchasing_gender_df["Avg Total Purchase per Person"] = purchasing_gender_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
purchasing_gender_df
```

```python
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table

# Create Bins for Age Buckets
bins = [0,9,14,19,24,29,34,39,max(purchase_data["Age"])]

# Lable the bins in order
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Add Age Buckets column based on bins
purchase_data["Age Bucket"] = pd.cut(purchase_data["Age"], bins=bins, labels=age_labels)

# Get count of screen names in each age bucket
less_ten_df = purchase_data.loc[purchase_data["Age Bucket"] == "<10",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
less_ten_count = less_ten_df.loc["<10","SN"]

ten_fourteen_df = purchase_data.loc[purchase_data["Age Bucket"] == "10-14",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
ten_fourteen_count = ten_fourteen_df.loc["10-14","SN"]
ten_fourteen_df

fifteen_nineteen_df = purchase_data.loc[purchase_data["Age Bucket"] == "15-19",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
fifteen_nineteen_count = fifteen_nineteen_df.loc["15-19","SN"]

twenty_twentyfour_df = purchase_data.loc[purchase_data["Age Bucket"] == "20-24",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
twenty_twentyfour_count = twenty_twentyfour_df.loc["20-24","SN"]

twentyfive_twentynine_df = purchase_data.loc[purchase_data["Age Bucket"] == "25-29",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
twentyfive_twentynine_count = twentyfive_twentynine_df.loc["25-29","SN"]

thirty_thirtyfour_df = purchase_data.loc[purchase_data["Age Bucket"] == "30-34",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
thirty_thirtyfour_count = thirty_thirtyfour_df.loc["30-34","SN"]

thirtyfive_thirtynine_df = purchase_data.loc[purchase_data["Age Bucket"] == "35-39",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
thirtyfive_thirtynine_count = thirtyfive_thirtynine_df.loc["35-39","SN"]

greater_forty_df = purchase_data.loc[purchase_data["Age Bucket"] == "40+",["Age Bucket","SN"]].groupby("Age Bucket").nunique("SN")
greater_forty_count = greater_forty_df.loc["40+","SN"]

# Create dataframe of data
age_bin_summary_df = pd.DataFrame({"Total Counts": [less_ten_count, 
                                                    ten_fourteen_count, 
                                                    fifteen_nineteen_count, 
                                                    twenty_twentyfour_count, 
                                                    twentyfive_twentynine_count, 
                                                    thirty_thirtyfour_count, 
                                                    thirtyfive_thirtynine_count, 
                                                    greater_forty_count],
                           "Percentage of Players": [less_ten_count/count_players*100, 
                                                     ten_fourteen_count/count_players*100, 
                                                     fifteen_nineteen_count/count_players*100,
                                                     twenty_twentyfour_count/count_players*100,
                                                     twentyfive_twentynine_count/count_players*100,
                                                     thirty_thirtyfour_count/count_players*100,
                                                     thirtyfive_thirtynine_count/count_players*100,
                                                     greater_forty_count/count_players*100
                                                    ]}, index = age_labels)

# Format data to show percentages
age_bin_summary_df["Percentage of Players"] = age_bin_summary_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

age_bin_summary_df

```

```python
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

# Sum Price by Age Bucket in small dataframe
sum_price_age_df = purchase_data[["Age Bucket", "Price"]].groupby("Age Bucket").sum()

# Count purchase IDs by age bucket
count_purchase_age_df = purchase_data[["Age Bucket", "Purchase ID"]].groupby("Age Bucket").count()

# Return aveage price per purchase by age bucket
average_price_age_df = purchase_data[["Age Bucket", "Price"]].groupby("Age Bucket").mean()

# Return unique count of screen names per age bucket
count_players_age_df = purchase_data[["Age Bucket", "SN"]].groupby("Age Bucket").nunique()

# Drop age bucket column
count_players2_age_df = count_players_age_df[["SN"]]

# combine all small dataframes
age_summary1_df = pd.merge(sum_price_age_df, count_purchase_age_df, how='inner', on=['Age Bucket'])
age_summary2_df = pd.merge(age_summary1_df, average_price_age_df, how="inner", on = ['Age Bucket'])
age_summary_df = pd.merge(age_summary2_df, count_players2_age_df, how= 'inner', on = ["Age Bucket"])

# Create new column for average purchase price per person
age_summary_df["Avg Total Purchase per Person"] = age_summary_df["Price_x"]/age_summary_df["SN"]

# Keep only necessary columns
purchasing_age_df = age_summary_df[["Purchase ID", "Price_y", "Price_x", "Avg Total Purchase per Person"]]

# Rename columns for better visuals
purchasing_age_df = purchasing_age_df.rename(columns = {"Purchase ID":"Purchase Count",
                                                              "Price_y": "Average Purchase Price",
                                                              "Price_x": "Total Purchase Value"})

# Reformat columns to show currency
purchasing_age_df["Average Purchase Price"] = purchasing_age_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchasing_age_df["Total Purchase Value"] = purchasing_age_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchasing_age_df["Avg Total Purchase per Person"] = purchasing_age_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

# Display dataframe
purchasing_age_df
```

```python
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

# Sum Price by screen name in small dataframe
sum_price_sn_df = purchase_data[["SN", "Price"]].groupby("SN").sum()

# Count purchase IDs by screen name
count_purchase_sn_df = purchase_data[["SN", "Purchase ID"]].groupby("SN").count()

# Return aveage price per purchase by screen name
average_price_sn_df = purchase_data[["SN", "Price"]].groupby("SN").mean()

# combine all small dataframes
sn_summary1_df = pd.merge(sum_price_sn_df, count_purchase_sn_df, how='inner', on=['SN'])
sn_summary_df = pd.merge(sn_summary1_df, average_price_sn_df, how="inner", on = ['SN'])

# Keep only necessary columns
purchasing_sn_df = sn_summary_df[["Purchase ID", "Price_y", "Price_x"]]

# Rename columns for better visuals
purchasing_sn_df = purchasing_sn_df.rename(columns = {"Purchase ID":"Purchase Count",
                                                              "Price_y": "Average Purchase Price",
                                                              "Price_x": "Total Purchase Value"})

# create sorted Dataframe
sorted_sn_df = purchasing_sn_df.sort_values("Total Purchase Value", ascending=False)

# reformat columns to show currency
sorted_sn_df["Average Purchase Price"] = sorted_sn_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
sorted_sn_df["Total Purchase Value"] = sorted_sn_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display dataframe summary
sorted_sn_df.head()

```

```python
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

# sum price by item variables in small dataframe
sum_price_item_df = purchase_data[["Item ID", "Item Name", "Price"]].groupby(["Item ID", "Item Name"]).sum()

# count purchases by Item variables in small dataframe
count_purchase_item_df = purchase_data[["Item ID", "Item Name", "Purchase ID"]].groupby(["Item ID", "Item Name"]).count()

# get price of each item in a small dataframe
max_price_item_df = purchase_data[["Item ID", "Item Name", "Price"]].groupby(["Item ID", "Item Name"]).max()

# Combine all small dataframs
item_summary1_df = pd.merge(sum_price_item_df, count_purchase_item_df, how='inner', on=["Item ID", "Item Name"])
item_summary_df = pd.merge(item_summary1_df, max_price_item_df, how="inner", on = ["Item ID", "Item Name"])

# Reorder columns
top_item_df = item_summary_df[["Purchase ID", "Price_y", "Price_x"]]

# Rename columns for better visuals
top_item_df = top_item_df.rename(columns = {"Purchase ID":"Purchase Count",
                                            "Price_y": "Item Price",
                                            "Price_x": "Total Purchase Value"})

# sort on popularity 
sorted_item_df = top_item_df.sort_values("Purchase Count", ascending=False)

# format columns of datafram to show currency
sorted_item_df["Item Price"] = sorted_item_df["Item Price"].astype(float).map("${:,.2f}".format)
sorted_item_df["Total Purchase Value"] = sorted_item_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display summary table
sorted_item_df.head()
```

```python
# Sort the above table by total purchase value in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the data frame

# Sort previous chart on Total purchase value
sorted_item_df = top_item_df.sort_values("Total Purchase Value", ascending=False)

# reformat data to show currency
sorted_item_df["Item Price"] = sorted_item_df["Item Price"].astype(float).map("${:,.2f}".format)
sorted_item_df["Total Purchase Value"] = sorted_item_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# print summary dataframe
sorted_item_df.head()
```
